## Maior idade


Crie um programa onde será perguntado o nome e a idade das pessoas e você irá seprar eles por Maior idade (mais que 18) e menor idade (menos que 18) (sem um limite pré-definido)

A pessoa pode:

- Adicionar Informação

- Remover Informação

- Consultar Informação

- Parar o Programa

Quando a pessoa solicitar excluir ela vai poder excluir através de uma dessas opções:

- Através do Nome

- Através da Idade

Além disso mostre para ele a lista de nomes e idades para ele saber qual deseja excluir

E mostrar para ela a informação que foi excluida, exemplo:

O usuário Carlos Vinicius que tinha 21 anos foi excluido com sucesso

Quando a pessoa solicitar consultar ela vai poder consultar através de uma dessas opções:

- Através do Nome

- Através da Idade

Quando a pessoa adicionar o nome e a idade:

Caso a idade da pessoa for maior do que 65 anos, ela será adicionada ao inicio da lista da maior idade

Verifique se a pessoa está digitando uma idade válida (Entre 0 e 125 anos)

- Mostre um texto falando "pessoas maiores de idade"
  - O nome da pessoa e a idade dela
